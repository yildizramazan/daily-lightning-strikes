#### convert the `Date Joined` column to datetime


```python
companies["Date Joined"] = pd.to_datetime(companies["Date Joined"])

companies.dtypes
```

#### creating a `Month Joined` column

```python
companies['Month Joined'] = companies["Date Joined"].dt.month_name()
companies.head()
```

#### creating a `Years To Join` column

```python
companies['Years To Join'] = companies["Date Joined"].dt.year - companies["Year Founded"]
companies.head(30)
```

#### gain more insight on a specific year

```python
companies_2021 = companies[companies["Date Joined"].dt.year == 2021].copy()

companies_2021.head()
```

```python
companies_2021.info()
```

#### observing trends over time

```python
companies_2021["Quarter Joined"] = companies_2021["Date Joined"].dt.quarter
```

```python
companies_2021.head()
```

```python
companies_2021["Quarter Joined"] = companies_2021["Date Joined"].dt.to_period("Q").dt.strftime("Q%q")
```

```python
companies_2021.head()
```

```python
companies_by_quarter_2021= companies_2021.groupby("Quarter Joined").agg(
    quarter_count = pd.NamedAgg(column="Quarter Joined", aggfunc="count"),
)
companies_by_quarter_2021.head()
```

```python
companies_2021.insert(3, "Week Joined", companies_2021["Date Joined"].dt.strftime('%Y-W%V'), True)

companies_by_week_2021 = companies_2021.groupby(by="Week Joined")["Company"].count().reset_index().rename(columns={"Company": "Company Count"})

companies_by_week_2021.head()
```

```python
companies_2020 = companies[companies["Date Joined"].dt.year == 2020].copy()

companies_2020.insert(3, "Week Joined", companies_2020["Date Joined"].dt.strftime('%Y-W%V'), True)

companies_by_week_2020 = companies_2020.groupby(by="Week Joined")["Company"].count().reset_index().rename(
    columns={"Company": "Company Count"})

companies_by_week_2020
```

```python
companies_by_week_20_21= pd.concat([companies_by_week_2021, companies_by_week_2020], axis=0, ignore_index=True)
```

```python
companies_by_week_20_21["Week Number"] = companies_by_week_20_21["Week Joined"].str[-2:]
```

```python
companies_by_week_20_21.head()
```

#### comparing trends over time

```python
companies_2020_2021 = pd.concat([companies_2020, companies_2021.drop(columns=['Week Joined', 'Quarter Joined'])])
```

```python
companies_2020_2021.head()
```

```python
companies_2020_2021["Quarter Joined"] = companies_2020_2021["Date Joined"].dt.to_period('Q').dt.strftime('%Y-Q%q')
```

```python
companies_2020_2021.columns
```

```python
companies_2020_2021["Valuation"] = companies_2020_2021["Valuation"].str.strip("$B").astype(float)
```

```python
companies_2020_2021["Valuation"].head()
```

```python
companies_by_quarter_2020_2021 = companies_2020_2021.groupby(by="Quarter Joined")["Valuation"].mean().reset_index().rename(columns={"Valuation":"Average Valuation"})
companies_by_quarter_2020_2021.head()
```

```python
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November",
               "December"]
```

```python
sns.boxplot(
    data=companies,
    x="Month Joined",
    y="Years To Join",
    order=month_order,
    showfliers=False,
    hue="Month Joined",

)

plt.title('Distribution of years to become unicorn with respect to month joined')
plt.xticks(rotation=45, horizontalalignment='right')
plt.show()

```

```python
plt.figure(figsize=(10,6));


sns.barplot(
    x=companies["Year Founded"],
    y=companies["Years To Join"],
    errorbar=('ci', False),

);

plt.xlabel('Years To Join Unicorn Status');

plt.xticks(rotation=45, horizontalalignment='right');

plt.show();
```

```python
companies_by_quarter_2020_2021["Year"] = companies_by_quarter_2020_2021["Quarter Joined"].str[:4]
companies_by_quarter_2020_2021["Quarter Number"] = companies_by_quarter_2020_2021["Quarter Joined"].str[-2:]


companies_by_quarter_2020_2021.head()
```

```python
plt.figure(figsize=(10,6));


sns.barplot(
    x=companies_by_quarter_2020_2021["Quarter Number"],
    y=companies_by_quarter_2020_2021["Average Valuation"],
    hue=companies_by_quarter_2020_2021["Year"],
)

plt.xlabel("Quarter")
plt.ylabel("Average Valuation")

plt.show();
```
