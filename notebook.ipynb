```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("home/jovyan/work/eda_using_basic_data_functions_in_python_dataset1.csv")
df.head(10)
```

```python
df["date"] = pd.to_datetime(df["date"])
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df['date']= pd.to_datetime(df['date'])
```

```python
# Calculate days with most lightning strikes.
df.groupby(['date']).sum().sort_values('number_of_strikes', ascending=False).head(10)
```

```python
df["month"] = df["date"].dt.month
df.head(10)
```

```python
# Calculate total number of strikes per month
df.groupby(['month'])[['number_of_strikes']].sum().sort_values('number_of_strikes', ascending=False).head(12)
```

```python
df["month_txt"] = df["date"].dt.month_name().str.slice(stop=3)
df.head(10)
```

```python
df_by_month = df.groupby(["month", "month_txt"])[["number_of_strikes"]].sum().sort_values(by=["month"], ascending=True).head(12).reset_index()
df_by_month.head(10)
```

```python
plt.bar(x=df_by_month['month_txt'],height= df_by_month['number_of_strikes'], label="Number of strikes")
plt.plot()

plt.xlabel("Months(2018)")
plt.ylabel("Number of lightning strikes")
plt.title("Number of lightning strikes in 2018 by months")
plt.legend()
plt.show()
```

```python
companies_sampled = companies.sample(random_state=5,n=50)
```

```python
companies_sampled
```

```python
companies_sampled["time_took"] = companies_sampled["Year Joined"] - companies_sampled["Year Founded"]
companies_by_time = companies_sampled[["Industry", "time_took"]].groupby(["Industry"]).max().sort_values(by=["time_took"])
companies_by_time
```

```python
print(companies_by_time.info())
plt.bar(x=companies_by_time.index,height= companies_by_time['time_took'])
plt.plot()

plt.xlabel("Industry")
plt.ylabel("Maximum number of years")
plt.title("Number of lightning strikes in 2018 by months")
plt.xticks(rotation=45, horizontalalignment='right')

plt.show()
```
